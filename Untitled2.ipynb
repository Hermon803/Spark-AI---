```python
import os
import json
import numpy as np
import torch
import torch.nn as nn
from torchvision import transforms, models
import matplotlib.pyplot as plt
import pydicom
import SimpleITK as sitk
from collections import defaultdict
import warnings
warnings.filterwarnings("ignore", category=UserWarning)  # 忽略PyTorch的警告

# 支持中文
plt.rcParams['font.sans-serif'] = ['SimHei'] # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False # 用来正常显示负号

# 配置参数
class Config:
    patch_size = 64
    num_vertebra_classes = 2  # 正常, 退行性病变
    num_disc_classes = 5      # 正常, 膨出, 突出, 脱出, 椎体内疝出
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model_path = r"D:\Progress\MRI\output\best_model.pth"  # 训练好的模型路径
    output_dir = r"D:\Progress\MRI\output"
    os.makedirs(output_dir, exist_ok=True)

# 双分支分类模型（与训练时一致）
class SpinalNet(nn.Module):
    def __init__(self, num_vertebra_classes, num_disc_classes):
        super(SpinalNet, self).__init__()
        try:
            backbone = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
        except AttributeError:
            backbone = models.resnet18(pretrained=True)
            
        backbone.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
        
        self.feature_extractor = nn.Sequential(
            backbone.conv1,
            backbone.bn1,
            backbone.relu,
            backbone.maxpool,
            backbone.layer1,
            backbone.layer2,
            backbone.layer3,
            backbone.layer4,
            backbone.avgpool
        )
        
        self.vertebra_classifier = nn.Sequential(
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, num_vertebra_classes)
        )
        
        self.disc_classifier = nn.Sequential(
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(512, num_disc_classes))
    
    def forward(self, x, label_types):
        features = self.feature_extractor(x)
        features = features.view(features.size(0), -1)
        
        outputs = torch.zeros(x.size(0), max(
            self.vertebra_classifier[-1].out_features,
            self.disc_classifier[-1].out_features
        )).to(x.device)
        
        vertebra_mask = (label_types == 0)
        if torch.any(vertebra_mask):
            vertebra_features = features[vertebra_mask]
            vertebra_output = self.vertebra_classifier(vertebra_features)
            outputs[vertebra_mask, :vertebra_output.size(1)] = vertebra_output
        
        disc_mask = (label_types == 1)
        if torch.any(disc_mask):
            disc_features = features[disc_mask]
            disc_output = self.disc_classifier(disc_features)
            outputs[disc_mask, :disc_output.size(1)] = disc_output
        
        return outputs

# 加载训练好的模型
def load_trained_model():
    model = SpinalNet(Config.num_vertebra_classes, Config.num_disc_classes)
    model.load_state_dict(torch.load(Config.model_path, map_location=Config.device))
    model.to(Config.device)
    model.eval()  # 设置为评估模式
    return model

# 关键点检测器（模拟版）
class KeypointDetector:
    def __init__(self, spacing=30):
        self.spacing = spacing  # 模拟关键点之间的间隔
    
    def detect_keypoints(self, image):
        """模拟关键点检测 - 在实际应用中应替换为真实的关键点检测算法"""
        height, width = image.shape
        points = []
        
        # 模拟检测脊柱关键点（从图像中间开始，每隔一定间距检测一个点）
        center_x = width // 2
        
        # 从图像高度1/4处开始，到3/4处结束
        start_y = height // 4
        end_y = height * 3 // 4
        
        # 生成关键点
        for y in range(start_y, end_y, self.spacing):
            # 随机偏移模拟真实检测
            offset_x = np.random.randint(-10, 10)
            points.append({
                "coord": [center_x + offset_x, y],
                "label_type": "disc" if np.random.rand() > 0.5 else "vertebra",
                "prob": 0.9  # 模拟检测置信度
            })
        
        return points

# 图像块提取函数
def extract_patch(img, x, y, patch_size=64):
    """提取以关键点为中心的图像块"""
    h, w = img.shape
    half_size = patch_size // 2
    
    # 计算边界
    x_min = max(0, int(x) - half_size)
    x_max = min(w, int(x) + half_size)
    y_min = max(0, int(y) - half_size)
    y_max = min(h, int(y) + half_size)
    
    # 提取图像块
    patch = img[y_min:y_max, x_min:x_max]
    
    # 填充
    pad_left = half_size - (int(x) - x_min)
    pad_right = half_size - (x_max - int(x))
    pad_top = half_size - (int(y) - y_min)
    pad_bottom = half_size - (y_max - int(y))
    
    patch = np.pad(patch, 
                  ((pad_top, pad_bottom), 
                   (pad_left, pad_right)), 
                  mode='constant', constant_values=0)
    
    # 归一化
    patch_min = patch.min()
    patch_max = patch.max()
    if patch_max > patch_min:
        patch = (patch - patch_min) / (patch_max - patch_min) * 255
    else:
        patch = np.zeros_like(patch)
    
    return patch.astype(np.uint8)

# 图像预处理转换
def get_transform(patch_size):
    return transforms.Compose([
        transforms.Resize((patch_size, patch_size)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5], std=[0.5])
    ])

# 判断DICOM图像是否为矢状位
import pydicom
import numpy as np
import SimpleITK as sitk
import math

def compute_orientation(vector):
    """
    根据方向向量计算DICOM图像方向标识符
    
    参数:
        vector (tuple): 包含三个浮点数的元组 (x, y, z)，表示方向向量
        
    返回:
        str: 图像方向标识符字符串，如"LPH"、"RAH"等
    """
    # 初始化方向字符串
    orientation = []
    
    # 解包向量分量
    x, y, z = vector
    
    # 确定每个分量的方向标识符
    # X轴: 负值为右(R)，正值为左(L)
    orientation_x = 'R' if x < 0 else 'L'
    
    # Y轴: 负值为前(A)，正值为后(P)
    orientation_y = 'A' if y < 0 else 'P'
    
    # Z轴: 负值为足(F)，正值为头(H)
    orientation_z = 'F' if z < 0 else 'H'
    
    # 计算各分量的绝对值
    abs_x = abs(x)
    abs_y = abs(y)
    abs_z = abs(z)
    
    # 设置比较阈值，避免浮点数精度问题
    epsilon = 0.0001
    
    # 按照重要性顺序添加方向标识符（从最大分量到最小分量）
    # 最多添加三个字符
    for _ in range(3):
        # 检查X分量是否是最大且超过阈值
        if abs_x > epsilon and abs_x > abs_y and abs_x > abs_z:
            orientation.append(orientation_x)
            abs_x = 0  # 设置为0，避免重复选择
        
        # 检查Y分量是否是最大且超过阈值
        elif abs_y > epsilon and abs_y > abs_x and abs_y > abs_z:
            orientation.append(orientation_y)
            abs_y = 0  # 设置为0，避免重复选择
        
        # 检查Z分量是否是最大且超过阈值
        elif abs_z > epsilon and abs_z > abs_x and abs_z > abs_y:
            orientation.append(orientation_z)
            abs_z = 0  # 设置为0，避免重复选择
        
        # 如果没有分量超过阈值，提前结束
        else:
            break
    
    # 将列表转换为字符串并返回
    return ''.join(orientation)

def is_sagittal_mri(dcm_path):
    """
    判断DICOM图像是否为矢状位MRI
    依据:
    1. 扫描方向(Scanning Sequence)中包含SE(自旋回波)或TSE(快速自旋回波)
    2. 序列名称(Protocol Name)中包含"sag"或"sagittal"
    3. 图像方向(ImageOrientationPatient)判断 - 使用compute_orientation函数
    4. 图像尺寸比例判断
    5. 序列描述(SeriesDescription)判断
    6. 图像类型(ImageType)判断
    """
    try:
        # 使用pydicom读取元数据
        ds = pydicom.dcmread(dcm_path, stop_before_pixels=True)
        
        # 1. 检查扫描序列
        scanning_sequence = getattr(ds, 'ScanningSequence', '').lower()
        if 'se' in scanning_sequence or 'tse' in scanning_sequence:
            return True
        
        # 2. 检查协议名称
        protocol_name = getattr(ds, 'ProtocolName', '').lower()
        if 'sag' in protocol_name or 'sagittal' in protocol_name:
            return True
        
        # 3. 使用compute_orientation函数改进图像方向判断
        if hasattr(ds, 'ImageOrientationPatient'):
            orientation = ds.ImageOrientationPatient
            
            # 提取行方向和列方向向量
            row_vector = orientation[0:3]
            col_vector = orientation[3:6]
            
            # 计算行方向和列方向的方向标识符
            row_orientation = compute_orientation(row_vector)
            col_orientation = compute_orientation(col_vector)
            
            # 计算法线方向（行方向和列方向的叉积）
            row = np.array(row_vector)
            col = np.array(col_vector)
            normal_vector = np.cross(row, col)
            normal_orientation = compute_orientation(normal_vector)
            
            # 矢状位的特征：法线方向为左右方向（L或R）
            # 行方向为前后方向（A或P），列方向为头足方向（H或F）
            if ('L' in normal_orientation or 'R' in normal_orientation) and \
               ('A' in row_orientation or 'P' in row_orientation) and \
               ('H' in col_orientation or 'F' in col_orientation):
                return True
        
        # 4. 检查图像尺寸比例
        if hasattr(ds, 'Rows') and hasattr(ds, 'Columns'):
            rows = ds.Rows
            cols = ds.Columns
            # 矢状位通常高度大于宽度
            if rows > cols * 1.5:
                return True
        
        # 5. 使用SimpleITK读取图像并分析形状
        reader = sitk.ImageFileReader()
        reader.SetFileName(dcm_path)
        reader.LoadPrivateTagsOn()
        reader.ReadImageInformation()
        
        size = reader.GetSize()
        # 矢状位通常高度大于宽度
        if len(size) >= 2 and size[0] < size[1] * 0.7:
            return True
        
        # 6. 检查序列描述
        series_description = getattr(ds, 'SeriesDescription', '').lower()
        if 'sag' in series_description or 'sagittal' in series_description:
            return True
        
        # 7. 检查图像类型
        image_type = getattr(ds, 'ImageType', [])
        image_type_str = ''.join(image_type).lower()
        if 'sag' in image_type_str or 'sagittal' in image_type_str:
            return True
        
        return False
    
    except Exception as e:
        print(f"判断图像方向时出错: {str(e)}")
        return False


# 预测单个MRI图像
def predict_mri(model, dcm_path, detector, transform):
    """预测单个MRI图像的病变类型"""
    try:
        # 读取DICOM图像
        reader = sitk.ImageFileReader()
        reader.SetImageIO("GDCMImageIO")
        reader.SetFileName(dcm_path)
        reader.LoadPrivateTagsOn()
        image = reader.Execute()
        img_array = sitk.GetArrayFromImage(image)[0]
        
        # 获取像素间距
        if reader.HasMetaDataKey("0028|0030"):
            pixel_spacing = [float(x) for x in reader.GetMetaData("0028|0030").split('\\')]
        else:
            pixel_spacing = [1.0, 1.0]
    except Exception as e:
        print(f"读取DICOM文件出错: {str(e)}")
        return None, None
    
    # 检测关键点
    keypoints = detector.detect_keypoints(img_array)
    
    # 准备图像块和标签类型
    patches = []
    label_types = []
    coords = []
    
    for kp in keypoints:
        x, y = kp["coord"]
        patch = extract_patch(img_array, x, y, Config.patch_size)
        patch_img = Image.fromarray(patch)
        patch_tensor = transform(patch_img)
        
        patches.append(patch_tensor)
        label_types.append(0 if kp["label_type"] == "vertebra" else 1)
        coords.append(kp["coord"])
    
    if not patches:
        print("未检测到关键点，无法进行分类预测。")
        return None, None
    
    # 转换为批量张量
    patches_tensor = torch.stack(patches).to(Config.device)
    label_types_tensor = torch.tensor(label_types, dtype=torch.long).to(Config.device)
    
    # 进行预测
    with torch.no_grad():
        outputs = model(patches_tensor, label_types_tensor)
        probs = torch.softmax(outputs, dim=1)
        pred_classes = torch.argmax(probs, dim=1).cpu().numpy()
        confidences = torch.max(probs, dim=1).values.cpu().numpy()
    
    # 整理预测结果
    results = []
    class_names = ["正常椎体", "退行性病变椎体", 
                  "正常椎间盘", "膨出", "突出", "脱出", "椎体内疝出"]
    
    for i, kp in enumerate(keypoints):
        # 确定类别索引
        if kp["label_type"] == "vertebra":
            class_idx = pred_classes[i]
            class_name = class_names[class_idx]
        else:
            class_idx = pred_classes[i] + 2  # 椎间盘类别偏移
            class_name = class_names[class_idx]
        
        results.append({
            "coord": kp["coord"],
            "label_type": kp["label_type"],
            "class_id": class_idx,
            "class_name": class_name,
            "confidence": confidences[i],
            "pixel_spacing": pixel_spacing
        })
    
    return img_array, results

# 可视化预测结果
def visualize_prediction(image, results, save_path=None):
    """可视化预测结果"""
    plt.figure(figsize=(12, 8))
    plt.imshow(image, cmap='gray')
    
    # 颜色映射
    colors = {
        "正常椎体": 'green',
        "退行性病变椎体": 'red',
        "正常椎间盘": 'blue', 
        
        "膨出": 'yellow',
        "突出": 'orange',
        "脱出": 'purple',
        "椎体内疝出": 'cyan'
    }
    
    # 绘制关键点和分类结果
    for result in results:
        x, y = result["coord"]
        plt.scatter(x, y, s=50, c=colors[result["class_name"]], marker='o', alpha=0.7)
        
        # 添加文本标签
        plt.text(x + 10, y, f"{result['class_name']} ({result['confidence']*100:.1f}%)", 
                color=colors[result["class_name"]], 
                fontsize=9, 
                bbox=dict(facecolor='black', alpha=0.5))
    
    plt.title("脊柱病变分类结果")
    plt.axis('off')
    
    if save_path:
        plt.savefig(save_path, bbox_inches='tight', dpi=300)
        print(f"结果图像已保存至: {save_path}")
    
    plt.show()

# 生成预测报告
def generate_report(results, output_path):
    """生成详细的预测报告"""
    # 按类型统计
    type_counts = defaultdict(int)
    class_counts = defaultdict(int)
    
    # 收集详细信息
    details = []
    
    for i, result in enumerate(results, 1):
        type_counts[result["label_type"]] += 1
        class_counts[result["class_name"]] += 1
        
        details.append({
            "序号": i,
            "位置": f"({result['coord'][0]:.1f}, {result['coord'][1]:.1f})",
            "类型": "椎体" if result["label_type"] == "vertebra" else "椎间盘",
            "分类结果": result["class_name"],
            "置信度": f"{result['confidence']*100:.1f}%"
        })
    
    # 创建报告
    report = {
        "总体统计": {
            "总关键点数": len(results),
            "椎体数量": type_counts.get("vertebra", 0),
            "椎间盘数量": type_counts.get("disc", 0),
            "分类分布": dict(class_counts)
        },
        "详细结果": details
    }
    
    # 保存为JSON
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(report, f, ensure_ascii=False, indent=2)
    
    print(f"预测报告已保存至: {output_path}")
    return report

# 主函数：加载模型并进行预测
def main():
    # 1. 加载训练好的模型
    print("加载预训练模型...")
    model = load_trained_model()
    print("模型加载完成!")
    
    # 2. 初始化关键点检测器和图像变换
    detector = KeypointDetector(spacing=40)
    transform = get_transform(Config.patch_size)
    
    # 3. 设置输入和输出路径
    input_dcm = r"D:\Progress\MRI\spinaldisease_lumbartestB50_datasets\study251\image10.dcm"  # 替换为你的DCM文件路径
    output_image = os.path.join(Config.output_dir, "prediction_result.png")
    output_report = os.path.join(Config.output_dir, "prediction_report.json")
    
    # 4. 检查是否为矢状位MRI
    print(f"检查图像方向: {input_dcm}")
    if not is_sagittal_mri(input_dcm):
        print("错误: 输入图像不是矢状位MRI图像，无法进行处理。")
        print("请提供矢状位T2加权脊柱MRI图像。")
        return
    
    print("图像为矢状位MRI，开始处理...")
    
    # 5. 进行预测
    image_array, results = predict_mri(model, input_dcm, detector, transform)
    
    if results:
        print(f"检测到 {len(results)} 个关键点，分类完成!")
        
        # 6. 可视化结果
        visualize_prediction(image_array, results, save_path=output_image)
        
        # 7. 生成详细报告
        report = generate_report(results, output_report)
        
        # 8. 打印摘要
        print("\n预测结果摘要:")
        print(f"总关键点数: {len(results)}")
        print(f"椎体数量: {report['总体统计']['椎体数量']}")
        print(f"椎间盘数量: {report['总体统计']['椎间盘数量']}")
        print("分类分布:")
        for cls, count in report['总体统计']['分类分布'].items():
            print(f"  {cls}: {count}")
    else:
        print("未检测到关键点或预测失败。")

if __name__ == "__main__":
    main()
```
